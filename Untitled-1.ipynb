```python
file = open("allIssues.txt", "r")
raw = file.read()
years = raw.split("\n\n")
```

```python
yearsDict = {}
filterPhrases = ["\tTable of Contents", "•\tIssue", "A Conversation with", "Book Reviews", "Books Reviews", "Books Received", "Books Review", "Editor's Note", "Editors' Note", "From the Editor's", "Recent Publications", "Interview with", "Author Index"]

for i, item in enumerate(years):
    itemDict = {}
    if (i % 2) == 0:
        strippedItem = item.split("(")[1].split(")")[0]
        yearsDict[strippedItem] = {}
        last_key = strippedItem
    else:
        issues = item.split("-\u2028Download PDF\u2028\u2028\n\t")[1:]
        yearsDict[last_key] = [item.split("\xa0")[0].split("\t")[1].split("\n")[0] for item in issues if len([phrase for phrase in filterPhrases if phrase in item]) == 0]
```

```python
import spacy
from spacy.language import Language

@Language.component("merge_contractions")
def merge_contractions(doc):
    spans = []
    for token in doc:
        # Merge contractions like 's, 've, 'll, 'd with preceding token
        if token.pos_ == "AUX" and token.text.startswith("'") and token.i > 0:
            spans.append((token.i - 1, token.i + 1))
        # Merge n't with the preceding token (e.g., Ca + n't -> Can't)
        elif token.text == "n't" and token.i > 0:
            spans.append((token.i - 1, token.i + 1))

    spans.sort(key=lambda s: (s[0], -s[1]))
    filtered = []
    for span in spans:
        if not filtered or span[0] >= filtered[-1][1]:
            filtered.append(span)

    with doc.retokenize() as retokenizer:
        for start, end in filtered:
            retokenizer.merge(doc[start:end])

    return doc

@Language.component("merge_hyphens")
def merge_hyphens(doc):
    spans = []
    i = 0
    while i < len(doc):
        if doc[i].text in ("-", "–", "—"):
            has_space_before = i > 0 and doc[i - 1].idx + len(doc[i - 1].text) < doc[i].idx
            has_space_after = i < len(doc) - 1 and doc[i].idx + len(doc[i].text) < doc[i + 1].idx

            if not has_space_before and not has_space_after and i > 0 and i < len(doc) - 1:
                start = i - 1
                end = i + 1

                # Expand through full chains: word-word-word-word
                while end + 1 < len(doc) and doc[end + 1].text in ("-", "–", "—"):
                    no_space_to_hyphen = doc[end].idx + len(doc[end].text) == doc[end + 1].idx
                    if no_space_to_hyphen and end + 2 < len(doc):
                        no_space_after_hyphen = doc[end + 1].idx + len(doc[end + 1].text) == doc[end + 2].idx
                        if no_space_after_hyphen:
                            end = end + 2
                        else:
                            break
                    else:
                        break

                spans.append(doc[start:end + 1])
                i = end + 1
                continue
        i += 1

    filtered = filter_spans(spans)

    with doc.retokenize() as retokenizer:
        for span in filtered:
            retokenizer.merge(span)

    return doc

from spacy.util import filter_spans

@Language.component("merge_named_entities")
def merge_named_entities(doc):
    spans = []

    allowed_labels = {"GPE", "ORG", "NORP"}
    for ent in doc.ents:
        # Only merge entities that are 2-4 tokens long
        if ent.label_ in allowed_labels and 1 < len(ent) <= 4:
            spans.append(doc[ent.start:ent.end])

    CUSTOM_TERMS = {
        ("United", "States"),
        ("United", "Nations"),
        ("United", "Kingdom"),
        ("European", "Union"),
        ("World", "Trade", "Organization"),
        ("World", "Bank"),
        ("International", "Monetary", "Fund"),
        ("International", "Court"),
        ("Security", "Council"),
        ("General", "Assembly"),
        ("Human", "Rights"),
        ("Cold", "War"),
        ("Middle", "East"),
        ("South", "Korea"),
        ("North", "Korea"),
        ("South", "Africa"),
        ("Saudi", "Arabia"),
        ("Hong", "Kong"),
        ("Sri", "Lanka"),
        ("Costa", "Rica"),
        ("Puerto", "Rico"),
        ("Soviet", "Union"),
        ("Third", "World"),
    }

    for term in CUSTOM_TERMS:
        term_len = len(term)
        for i in range(len(doc) - term_len + 1):
            window = tuple(doc[i + j].text for j in range(term_len))
            if window == term:
                spans.append(doc[i:i + term_len])

    filtered = filter_spans(spans)

    with doc.retokenize() as retokenizer:
        for span in filtered:
            retokenizer.merge(span)

    return doc

nlp = spacy.load("en_core_web_sm")
nlp.add_pipe("merge_named_entities", after="ner")
nlp.add_pipe("merge_hyphens", after="ner")
nlp.add_pipe("merge_contractions", after="merge_hyphens")
print(nlp.pipe_names)
```

```python
def normalize_token(text):
    """Convert dot-separated acronyms (U.S. -> US) and lowercase everything else."""
    if re.match(r'^([A-Za-z]\.)+$', text):
        return text.replace('.', '').upper()
    return text.lower()
```
